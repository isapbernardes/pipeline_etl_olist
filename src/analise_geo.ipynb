```python
import pandas as pd

# Carrega o arquivo
df = pd.read_csv(r"C:\Users\isado\OneDrive\Área de Trabalho\projeto_etl\dados_brutos\olist_geolocation_dataset.csv")

print(f"Linhas totais originais: {len(df)}")

# 1. Duplicatas EXATAS (Linha inteira igual)
duplicatas_exatas = df.duplicated().sum()
print(f"Linhas idênticas (lixo puro): {duplicatas_exatas}")

# 2. Vamos pegar um exemplo de CEP para você ver com seus próprios olhos
cep_exemplo = 24220 # Um CEP de Niterói/RJ que costuma ter muitos dados
print(f"\n--- Analisando o CEP {cep_exemplo} ---")

df_cep = df[df['geolocation_zip_code_prefix'] == cep_exemplo]
print(f"Total de registros para esse CEP: {len(df_cep)}")
print("\nPrimeiras 5 linhas desse CEP (Veja se parecem repetidas):")
print(df_cep.head())

# 3. Teste do Drop
df_limpo = df.drop_duplicates()
print(f"\nLinhas após limpeza segura: {len(df_limpo)}")
```
